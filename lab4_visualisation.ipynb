```python
import random
from matplotlib import pyplot as plt

# Binary Tree Node
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

# Binary Tree Creation with random real numbers
def create_binary_tree(size, min_val=10, max_val=50):
    def insert(node, data):
        if node is None:
            return Node(data)
        elif data < node.data:
            node.left = insert(node.left, data)
        else:
            node.right = insert(node.right, data)
        return node

    root = None
    for _ in range(size):
        data = random.uniform(min_val, max_val)
        root = insert(root, data)
    return root

# Knuth's Algorithm for Tree Visualization
def visualize_binary_tree(node, x=0, y=0, dx=20, dy=20):
    if node:
        plt.text(x, y, f"{node.data:.2f}", ha="center", va="center")
        if node.left:
            x_left = x - dx
            y_left = y - dy
            plt.plot((x, x_left), (y, y_left), "k-")
            visualize_binary_tree(node.left, x_left, y_left, dx, dy)
        if node.right:
            x_right = x + dx
            y_right = y - dy
            plt.plot((x, x_right), (y, y_right), "k-")
            visualize_binary_tree(node.right, x_right, y_right, dx, dy)

# Create the binary tree
tree_size = 15
root = create_binary_tree(tree_size)

# Visualization using Knuth's algorithm
plt.figure(figsize=(12, 8))
visualize_binary_tree(root)
plt.title(f"Visualization of a Binary Tree (Knuth's Algorithm) with {tree_size} Nodes")
plt.axis("off")
plt.show()
```
